```python
import json
import requests
```

```python
jsonl_file = "C:/Users/ohs/Documents/legislation/hpws/legi_content.jsonl"
count_json = 0
count_hwp = 0
with open(jsonl_file, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        count_json += 1
        if data['법령안hwp'] is not None:
            count_hwp += 1
            if count_json != count_hwp:
                print(f"error at {i}")
                count_hwp, count_json = 0, 0
            url = data['법령안hwp']
            save_name = f"legi_{data['id']:>04d}.hwp"
            with open(save_name, 'wb') as f:
                f.write(requests.get(url).content)
```
